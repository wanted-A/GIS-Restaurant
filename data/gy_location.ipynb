### 경기도 위도 경도 코드

```python
import pandas as pd
import json

df = pd.read_csv("location.csv")
```

```python
gy = df.loc[(df["do-si"] == "경기")]
gy
```

```python
json_list = []

for index, row in gy.iterrows():
    data = {
        "model": "restaurants.Location",
        "pk": index-17,
        "fields": {
            "do_si": row['do-si'],
            "sgg": row['sgg'],
            "lon": str(row['lon']),
            "lat": str(row['lat'])
        }
    }
    
    json_list.append(data)
```

```python
with open("gy_location.json", "w", encoding="utf-8") as file:
    json.dump(json_list, file, ensure_ascii=False, indent="\t")
```
