### 대한민국 전체 위도 경도 코드

```python
import pandas as pd
import json

raw = pd.read_csv("location.csv")
raw
```

```python
json_list = []

for index, row in raw.iterrows():
    data = {
        "models": "raw.location",
        "pk" : index+1,
        "fields":{
            "do_si": row["do-si"],
            "sgg": row["sgg"],
            "lon": str(row["lon"]),
            "lat": str(row["lat"])
        }
    }

    json_list.append(data)
```

```python
with open("raw_location.json", "w") as json_file:
    json.dump(json_list, json_file, ensure_ascii=False, indent="\t")
```
