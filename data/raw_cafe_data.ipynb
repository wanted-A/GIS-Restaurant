### 전체 카페 정보 저장

```python
import os
import environ
import pandas as pd
import requests
import json

env = environ.Env()

CAFE_KEY = env("CAFE_KEY")

url = f"https://openapi.gg.go.kr/Genrestrtcate?Key={CAFE_KEY}&Type=json"
response = requests.get(url)
data = response.json()
data
```

```python
cafes = pd.DataFrame(data["Genrestrtcate"][1]["row"])
cafes
```

```python
# cafes.to_json("raw_cafe_data_temp.json", orient="records", lines=True, force_ascii=False)
```

```python
json_list = []

for index, row in cafes.iterrows():
    json_dict = {}
    json_dict["model"] = "cafes.cafe"
    json_dict["pk"] = index + 1
    json_dict["fields"] = row.to_dict()
    fields = row.to_dict()
    
    for key, value in fields.items():
        if pd.isna(value):
            fields[key] = None
    
    json_dict["fields"] = fields
    json_list.append(json_dict)
```

```python
with open("raw_cafe_data.json", "w", encoding="utf-8") as file:
    json.dump(json_list, file, ensure_ascii=False, indent="\t")
```
